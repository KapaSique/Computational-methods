# Задача 6: Численное интегрирование
---

### Задание

Посчитать интеграл функции $f(x) = e^x$ на отрезке $[0, 1]$ с помощью методов:
- Центральных прямоугольников
- Трапеций

Рассмотреть вариацию числа отрезков: 4, 8, 16, 32, 64.

Вывести значение погрешности для всех 10 приближенных значений интеграла.

### Точное значение

$$\int_0^1 e^x \, dx = e^x \Big|_0^1 = e - 1 \approx 1.718281828$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Функция для интегрирования
def f(x):
    return np.exp(x)

# Точное значение интеграла на [0, 1]
exact_value = np.e - 1
print(f"Точное значение интеграла: {exact_value:.15f}")
```

---
## Метод центральных прямоугольников

Формула:

$$\int_a^b f(x) \, dx \approx h \sum_{i=0}^{n-1} f\left(x_i + \frac{h}{2}\right)$$

где $h = \frac{b-a}{n}$, $x_i = a + i \cdot h$

```python
def midpoint_rule(f, a, b, n):
    """
    Метод центральных прямоугольников
    
    Args:
        f: функция для интегрирования
        a, b: границы интегрирования
        n: число отрезков
    """
    h = (b - a) / n
    result = 0.0
    
    for i in range(n):
        x_mid = a + (i + 0.5) * h
        result += f(x_mid)
    
    return h * result
```

---
## Метод трапеций

Формула:

$$\int_a^b f(x) \, dx \approx h \left[ \frac{f(a) + f(b)}{2} + \sum_{i=1}^{n-1} f(x_i) \right]$$

где $h = \frac{b-a}{n}$, $x_i = a + i \cdot h$

```python
def trapezoidal_rule(f, a, b, n):
    """
    Метод трапеций
    
    Args:
        f: функция для интегрирования
        a, b: границы интегрирования
        n: число отрезков
    """
    h = (b - a) / n
    result = (f(a) + f(b)) / 2.0
    
    for i in range(1, n):
        x_i = a + i * h
        result += f(x_i)
    
    return h * result
```

---
## Вычисление интеграла с различным числом отрезков

```python
# Границы интегрирования
a, b = 0, 1

# Число отрезков
n_values = [4, 8, 16, 32, 64]

# Результаты
results = []

print("=" * 80)
print("ЧИСЛЕННОЕ ИНТЕГРИРОВАНИЕ: f(x) = e^x на [0, 1]")
print("=" * 80)
print(f"Точное значение: {exact_value:.15f}\n")

print(f"{'Метод':<25} {'n':<8} {'Приближение':<20} {'Погрешность':<15}")
print("-" * 80)

for n in n_values:
    # Метод центральных прямоугольников
    mid_result = midpoint_rule(f, a, b, n)
    mid_error = abs(mid_result - exact_value)
    results.append(('Центр. прямоугольники', n, mid_result, mid_error))
    print(f"{'Центр. прямоугольники':<25} {n:<8} {mid_result:<20.15f} {mid_error:<15.2e}")
    
    # Метод трапеций
    trap_result = trapezoidal_rule(f, a, b, n)
    trap_error = abs(trap_result - exact_value)
    results.append(('Трапеции', n, trap_result, trap_error))
    print(f"{'Трапеции':<25} {n:<8} {trap_result:<20.15f} {trap_error:<15.2e}")
    print()

print("=" * 80)
```

---
## Визуализация сходимости

```python
# Извлечение данных для графиков
mid_errors = [r[3] for r in results if r[0] == 'Центр. прямоугольники']
trap_errors = [r[3] for r in results if r[0] == 'Трапеции']

# График погрешности в логарифмическом масштабе
plt.figure(figsize=(12, 5))

# Левый график: погрешность vs число отрезков
plt.subplot(1, 2, 1)
plt.plot(n_values, mid_errors, 'o-', label='Центральные прямоугольники', linewidth=2, markersize=8)
plt.plot(n_values, trap_errors, 's-', label='Трапеции', linewidth=2, markersize=8)
plt.xlabel('Число отрезков n', fontsize=12)
plt.ylabel('Абсолютная погрешность', fontsize=12)
plt.title('Сходимость методов численного интегрирования', fontsize=13)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.yscale('log')
plt.xscale('log')

# Правый график: погрешность vs шаг h
plt.subplot(1, 2, 2)
h_values = [(b - a) / n for n in n_values]
plt.plot(h_values, mid_errors, 'o-', label='Центральные прямоугольники', linewidth=2, markersize=8)
plt.plot(h_values, trap_errors, 's-', label='Трапеции', linewidth=2, markersize=8)

# Теоретические оценки O(h^2)
h_theory = np.array(h_values)
plt.plot(h_theory, 0.1 * h_theory**2, '--', color='gray', alpha=0.7, label='O(h²) (теория)')

plt.xlabel('Шаг h', fontsize=12)
plt.ylabel('Абсолютная погрешность', fontsize=12)
plt.title('Зависимость погрешности от шага', fontsize=13)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.yscale('log')
plt.xscale('log')

plt.tight_layout()
plt.show()
```

---
## Анализ результатов

### Наблюдения:

1. **Порядок сходимости**: Оба метода имеют порядок сходимости O(h²), что видно на графике в логарифмическом масштабе

2. **Точность методов**: 
   - Метод центральных прямоугольников обычно дает более точные результаты
   - Метод трапеций также показывает хорошую сходимость

3. **Уменьшение погрешности**: При увеличении числа отрезков в 2 раза погрешность уменьшается примерно в 4 раза (O(h²))

### Теоретические оценки погрешности:

**Метод центральных прямоугольников:**
$$E_{mid} = \frac{(b-a)^3}{24n^2} f''(\xi), \quad \xi \in [a,b]$$

**Метод трапеций:**
$$E_{trap} = -\frac{(b-a)^3}{12n^2} f''(\xi), \quad \xi \in [a,b]$$

Для $f(x) = e^x$: $f''(x) = e^x$, максимум на $[0,1]$ достигается при $x=1$: $f''(1) = e \approx 2.718$

## Визуализация методов интегрирования

```python
# Визуализация геометрического смысла методов
n_demo = 8
h = (b - a) / n_demo
x_plot = np.linspace(a, b, 200)
y_plot = f(x_plot)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Метод центральных прямоугольников
ax1.plot(x_plot, y_plot, 'b-', linewidth=2, label='f(x) = e^x')
for i in range(n_demo):
    x_left = a + i * h
    x_right = a + (i + 1) * h
    x_mid = (x_left + x_right) / 2
    y_mid = f(x_mid)
    ax1.add_patch(plt.Rectangle((x_left, 0), h, y_mid, 
                                 fill=True, alpha=0.3, edgecolor='red', facecolor='red'))
    ax1.plot([x_mid], [y_mid], 'ro', markersize=6)

ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('f(x)', fontsize=12)
ax1.set_title(f'Метод центральных прямоугольников (n={n_demo})', fontsize=13)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Метод трапеций
ax2.plot(x_plot, y_plot, 'b-', linewidth=2, label='f(x) = e^x')
x_nodes = np.linspace(a, b, n_demo + 1)
y_nodes = f(x_nodes)
for i in range(n_demo):
    x_trap = [x_nodes[i], x_nodes[i+1], x_nodes[i+1], x_nodes[i]]
    y_trap = [0, 0, y_nodes[i+1], y_nodes[i]]
    ax2.fill(x_trap, y_trap, alpha=0.3, edgecolor='green', facecolor='green')

ax2.plot(x_nodes, y_nodes, 'go-', markersize=6, linewidth=1.5)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('f(x)', fontsize=12)
ax2.set_title(f'Метод трапеций (n={n_demo})', fontsize=13)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
