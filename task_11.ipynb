# Задача: Проверка формулы Симпсона

Показать что погрешность интегрирования с использованием формулы Симпсона для функции x³ равна нулю, а для функции e^x не нулевая. Пределы интегрирования и количество отрезков разбиения можно выбрать любые.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Теория

Формула Симпсона (парабол) для численного интегрирования:

$$\int_a^b f(x)dx \approx \frac{h}{3}\left[f(x_0) + 4\sum_{i=1,3,5,...}^{n-1}f(x_i) + 2\sum_{i=2,4,6,...}^{n-2}f(x_i) + f(x_n)\right]$$

где $h = \frac{b-a}{n}$, $n$ - четное число отрезков.

**Важное свойство:** Формула Симпсона точна для полиномов степени не выше 3, т.е. погрешность равна нулю для функций вида $f(x) = ax^3 + bx^2 + cx + d$.

## Реализация формулы Симпсона

```python
def simpson_rule(f, a, b, n):
    """
    Численное интегрирование методом Симпсона.
    
    Parameters:
    f: функция для интегрирования
    a: нижний предел интегрирования
    b: верхний предел интегрирования
    n: количество отрезков разбиения (должно быть четным)
    
    Returns:
    Приближенное значение интеграла
    """
    if n % 2 != 0:
        raise ValueError("Количество отрезков должно быть четным")
    
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)
    
    # Формула Симпсона
    result = y[0] + y[-1]
    result += 4 * np.sum(y[1:-1:2])  # нечетные индексы
    result += 2 * np.sum(y[2:-1:2])  # четные индексы
    result *= h / 3
    
    return result
```

## Тест 1: Функция f(x) = x³

Для этой функции погрешность должна быть равна нулю (с точностью до машинной погрешности).

```python
# Определяем функцию f(x) = x^3
def f1(x):
    return x**3

# Пределы интегрирования
a1, b1 = 0, 2

# Аналитическое значение интеграла: ∫x³dx = x⁴/4
exact_1 = (b1**4 - a1**4) / 4

print(f"Точное значение интеграла ∫₀² x³ dx = {exact_1}")
print("\nРезультаты численного интегрирования методом Симпсона:")
print("-" * 60)

# Проверяем для разного количества отрезков
n_values = [2, 4, 6, 10, 20, 50]

for n in n_values:
    approx = simpson_rule(f1, a1, b1, n)
    error = abs(exact_1 - approx)
    print(f"n = {n:3d}: Приближение = {approx:.15f}, Погрешность = {error:.2e}")
```

## Тест 2: Функция f(x) = e^x

Для этой функции погрешность будет ненулевой, но будет уменьшаться с увеличением количества отрезков.

```python
# Определяем функцию f(x) = e^x
def f2(x):
    return np.exp(x)

# Пределы интегрирования
a2, b2 = 0, 1

# Аналитическое значение интеграла: ∫e^x dx = e^x
exact_2 = np.exp(b2) - np.exp(a2)

print(f"Точное значение интеграла ∫₀¹ e^x dx = {exact_2:.15f}")
print("\nРезультаты численного интегрирования методом Симпсона:")
print("-" * 60)

errors_exp = []
n_values_exp = [2, 4, 6, 10, 20, 50, 100]

for n in n_values_exp:
    approx = simpson_rule(f2, a2, b2, n)
    error = abs(exact_2 - approx)
    errors_exp.append(error)
    print(f"n = {n:3d}: Приближение = {approx:.15f}, Погрешность = {error:.2e}")
```

## Визуализация сходимости погрешности

```python
plt.figure(figsize=(12, 5))

# График погрешности для e^x
plt.subplot(1, 2, 1)
plt.loglog(n_values_exp, errors_exp, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Количество отрезков n', fontsize=12)
plt.ylabel('Абсолютная погрешность', fontsize=12)
plt.title('Погрешность для f(x) = e^x', fontsize=13)
plt.grid(True, alpha=0.3, which='both')

# Теоретическая оценка O(h^4) = O(n^-4)
h_values = (b2 - a2) / np.array(n_values_exp)
theoretical = 0.01 * h_values**4
plt.loglog(n_values_exp, theoretical, 'r--', linewidth=1.5, label='O(n⁻⁴)', alpha=0.7)
plt.legend(fontsize=10)

# Сравнение функций
plt.subplot(1, 2, 2)
x_plot = np.linspace(0, 2, 100)
plt.plot(x_plot, x_plot**3, 'b-', linewidth=2, label='f(x) = x³')
plt.plot(x_plot, np.exp(x_plot), 'r-', linewidth=2, label='f(x) = e^x')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Сравнение функций', fontsize=13)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Дополнительная проверка: другие полиномы степени ≤ 3

```python
# Проверим, что формула Симпсона точна для всех полиномов степени ≤ 3
print("Проверка точности для различных полиномов:")
print("=" * 70)

test_functions = [
    (lambda x: 1 + 0*x, "f(x) = 1", lambda a, b: b - a),
    (lambda x: x, "f(x) = x", lambda a, b: (b**2 - a**2) / 2),
    (lambda x: x**2, "f(x) = x²", lambda a, b: (b**3 - a**3) / 3),
    (lambda x: x**3, "f(x) = x³", lambda a, b: (b**4 - a**4) / 4),
    (lambda x: 2*x**3 - 3*x**2 + 5*x - 1, "f(x) = 2x³ - 3x² + 5x - 1", 
     lambda a, b: (b**4/2 - a**4/2) - (b**3 - a**3) + (5*b**2/2 - 5*a**2/2) - (b - a)),
]

a_test, b_test = 1, 3
n_test = 10

for func, name, exact_func in test_functions:
    exact = exact_func(a_test, b_test)
    approx = simpson_rule(func, a_test, b_test, n_test)
    error = abs(exact - approx)
    print(f"{name:30s}: Погрешность = {error:.2e}")
```

## Выводы

1. **Для функции f(x) = x³:** Погрешность интегрирования методом Симпсона равна нулю (с точностью до машинной погрешности ~10⁻¹⁵), что подтверждает теоретическое свойство формулы Симпсона.

2. **Для функции f(x) = e^x:** Погрешность ненулевая и уменьшается с увеличением количества отрезков разбиения. Скорость сходимости соответствует теоретической оценке O(h⁴) или O(n⁻⁴).

3. **Общий вывод:** Формула Симпсона точна для всех полиномов степени не выше 3, что делает её эффективным методом численного интегрирования для гладких функций.
