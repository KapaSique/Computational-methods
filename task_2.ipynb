```python
import math
import matplotlib.pyplot as plt

N = 1_000_000

PI_VARIANTS = [
    "3.14159265358979",
    "3.141592653589",
    "3.1415926535",
    "3.14159265",
    "3.141592",
    "3.1415",
    "3.14",
    "3",
]

def harmonic_2_fsum(n: int) -> float:
    return math.fsum(1.0 / (k * k) for k in range(1, n + 1))

def sum_by_recurrence(pi_value: float, n: int) -> float:
    s = 0.0
    for k in range(n, 0, -1):
        s += pi_value / (k * k)
    return s

def main():
    H2 = harmonic_2_fsum(N)
    S_true = math.pi * H2

    variants = []
    rel_errors = []

    # шапка таблицы
    print(f"S* (эталон) = {S_true:.16f}\n")
    print(f"{'Вар':>3}  {'pi(ввод)':>16}  {'S(расчёт)':>18}  {'Δ(S)':>14}  {'δ(S)':>14}")
    print("-" * 75)

    for i, p_str in enumerate(PI_VARIANTS, start=1):
        p = float(p_str)
        S = sum_by_recurrence(p, N)

        abs_err = abs(S_true - S)
        rel_err = abs_err / abs(S_true)

        variants.append(i)
        rel_errors.append(rel_err)

        print(f"{i:>3}  {p_str:>16}  {S:>18.12f}  {abs_err:>14.6e}  {rel_err:>14.6e}")

    # график
    plt.figure()
    plt.plot(variants, rel_errors, marker="o")
    plt.yscale("log")
    plt.xlabel("Вариант (1 — самое точное pi, 8 — pi=3)")
    plt.ylabel("Относительная погрешность δ(S)")
    plt.title("Зависимость δ(S) от точности pi")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()

if __name__ == "__main__":
    main()
```
