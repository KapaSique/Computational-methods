```python
import math
import matplotlib.pyplot as plt

N = 1_000_000

PI_VARIANTS = [
    "3.14159265358979",
    "3.141592653589",
    "3.1415926535",
    "3.14159265",
    "3.141592",
    "3.1415",
    "3.14",
    "3",
]

def harmonic_2_fsum(n: int) -> float:
    return math.fsum(1.0 / (k * k) for k in range(1, n + 1))

def sum_by_recurrence(pi_value: float, n: int) -> float:
    s = 0.0
    for k in range(n, 0, -1):
        s += pi_value / (k * k)
    return s

def main():
    H2 = harmonic_2_fsum(N)
    S_true = math.pi * H2

    variants = []
    rel_errors = []

    # шапка таблицы
    print(f"S* (эталон) = {S_true:.16f}\n")
    print(f"{'Вар':>3}  {'pi(ввод)':>16}  {'S(расчёт)':>18}  {'Δ(S)':>14}  {'δ(S)':>14}")
    print("-" * 75)

    for i, p_str in enumerate(PI_VARIANTS, start=1):
        p = float(p_str)
        S = sum_by_recurrence(p, N)

        abs_err = abs(S_true - S)
        rel_err = abs_err / abs(S_true)

        variants.append(i)
        rel_errors.append(rel_err)

        print(f"{i:>3}  {p_str:>16}  {S:>18.12f}  {abs_err:>14.6e}  {rel_err:>14.6e}")

    # график
    plt.figure()
    plt.plot(variants, rel_errors, marker="o")
    plt.yscale("log")
    plt.xlabel("Вариант (1 — самое точное pi, 8 — pi=3)")
    plt.ylabel("Относительная погрешность δ(S)")
    plt.title("Зависимость δ(S) от точности pi")
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import math
import matplotlib.pyplot as plt

N = 1_000_000

pi_values = [
    3.0,
    3.14,
    3.1415,
    3.141592,
    3.14159265,
    3.1415926535,
    3.141592653589,
    3.14159265358979,
]

def sum_recurrent(p: float, n_max: int) -> float:
    s = 0.0
    for n in range(n_max, 0, -1):          
        s += p / (n * n)
    return s

s_exact = sum_recurrent(math.pi, N)

rows = []
for p in pi_values:
    s = sum_recurrent(p, N)
    abs_err = abs(s_exact - s)
    rel_err = abs_err / abs(s_exact)
    rows.append((p, s, abs_err, rel_err))

header = f"{'pi':>16} {'S':>18} {'Δ(S)':>14} {'σ(S)':>14}"
print(header)
print("-" * len(header))
for p, s, abs_err, rel_err in rows:
    print(f"{p:>16.14g} {s:>18.12g} {abs_err:>14.6e} {rel_err:>14.6e}")

variants = list(range(1, len(pi_values) + 1))
rel_errors = [r[3] for r in rows]

plt.figure()
plt.plot(variants, rel_errors, marker="o")
plt.yscale("log")
plt.xlabel("Вариант (1 — pi=3, 8 — самое точное pi)")
plt.ylabel("Относительная погрешность σ(S)")
plt.title("Зависимость σ(S) от точности pi")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
