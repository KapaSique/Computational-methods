# Задача: Интерполяция Ньютона

Построить интерполяционные полиномы Ньютона для функции sin(x) на отрезке [0, 2] при использовании четырех и шести случайных точек. Результаты представить в виде графиков двух полиномов и синуса на заданном отрезке на одной координатой плоскости.

```python
import numpy as np
import matplotlib.pyplot as plt

# Устанавливаем seed для воспроизводимости
np.random.seed(42)
```

## Реализация метода Ньютона

Интерполяционный полином Ньютона использует разделенные разности:

$$P_n(x) = f[x_0] + f[x_0, x_1](x - x_0) + f[x_0, x_1, x_2](x - x_0)(x - x_1) + \ldots$$

```python
def divided_differences(x, y):
    """
    Вычисление разделенных разностей для интерполяции Ньютона.
    
    Parameters:
    x: массив узлов интерполяции
    y: массив значений функции в узлах
    
    Returns:
    Массив разделенных разностей
    """
    n = len(x)
    # Создаем таблицу разделенных разностей
    F = np.zeros((n, n))
    F[:, 0] = y
    
    for j in range(1, n):
        for i in range(n - j):
            F[i, j] = (F[i + 1, j - 1] - F[i, j - 1]) / (x[i + j] - x[i])
    
    # Возвращаем первую строку - коэффициенты полинома Ньютона
    return F[0, :]
```

```python
def newton_polynomial(x_nodes, y_nodes, x_eval):
    """
    Вычисление значений интерполяционного полинома Ньютона.
    
    Parameters:
    x_nodes: узлы интерполяции
    y_nodes: значения функции в узлах
    x_eval: точки, в которых нужно вычислить полином
    
    Returns:
    Значения полинома в точках x_eval
    """
    coef = divided_differences(x_nodes, y_nodes)
    n = len(x_nodes)
    
    # Вычисляем значения полинома по схеме Горнера
    result = np.zeros_like(x_eval)
    
    for i, x in enumerate(x_eval):
        p = coef[n - 1]
        for j in range(n - 2, -1, -1):
            p = p * (x - x_nodes[j]) + coef[j]
        result[i] = p
    
    return result
```

## Генерация случайных узлов и построение полиномов

```python
# Определяем функцию
f = np.sin

# Отрезок интерполяции
a, b = 0, 2

# Генерируем случайные узлы
x_4 = np.sort(np.random.uniform(a, b, 4))
x_6 = np.sort(np.random.uniform(a, b, 6))

# Вычисляем значения функции в узлах
y_4 = f(x_4)
y_6 = f(x_6)

print("Узлы для 4 точек:", x_4)
print("Узлы для 6 точек:", x_6)
```

```python
# Точки для построения графиков
x_plot = np.linspace(a, b, 200)

# Вычисляем значения полиномов Ньютона
y_newton_4 = newton_polynomial(x_4, y_4, x_plot)
y_newton_6 = newton_polynomial(x_6, y_6, x_plot)

# Истинные значения функции
y_true = f(x_plot)
```

## Визуализация результатов

```python
plt.figure(figsize=(12, 7))

# График истинной функции
plt.plot(x_plot, y_true, 'k-', linewidth=2, label='sin(x)', alpha=0.7)

# Полином Ньютона с 4 точками
plt.plot(x_plot, y_newton_4, 'b--', linewidth=1.5, label='Полином Ньютона (4 точки)')
plt.plot(x_4, y_4, 'bo', markersize=8, label='Узлы (4 точки)')

# Полином Ньютона с 6 точками
plt.plot(x_plot, y_newton_6, 'r--', linewidth=1.5, label='Полином Ньютона (6 точек)')
plt.plot(x_6, y_6, 'ro', markersize=8, label='Узлы (6 точек)')

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Интерполяция функции sin(x) полиномами Ньютона', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(a, b)

plt.tight_layout()
plt.show()
```

## Анализ погрешности

```python
# Вычисляем максимальные погрешности
error_4 = np.abs(y_true - y_newton_4)
error_6 = np.abs(y_true - y_newton_6)

print(f"Максимальная погрешность для 4 точек: {np.max(error_4):.6f}")
print(f"Максимальная погрешность для 6 точек: {np.max(error_6):.6f}")

# График погрешностей
plt.figure(figsize=(12, 5))

plt.plot(x_plot, error_4, 'b-', linewidth=1.5, label='Погрешность (4 точки)')
plt.plot(x_plot, error_6, 'r-', linewidth=1.5, label='Погрешность (6 точек)')

plt.xlabel('x', fontsize=12)
plt.ylabel('|sin(x) - P(x)|', fontsize=12)
plt.title('Погрешность интерполяции', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(a, b)

plt.tight_layout()
plt.show()
```

2. Полином с 6 точками обеспечивает более точное приближение функции sin(x)
3. Оба полинома точно проходят через заданные узлы интерполяции
4. С увеличением количества узлов погрешность интерполяции уменьшается
