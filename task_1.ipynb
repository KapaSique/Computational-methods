Задача 1: Вычислительная погрешность
Вычислить абсолютное и относительное значение погрешности расчета суммы

при использовании рекуррентной формулы:
S_n=S_(n-1)+π/n^2 ,n=1,…,1 000 000,S_0=0.
В качестве точной суммы принять решения при использовании рекуррентной формулы:
∑_(n-1)▒= ∑_n▒〖+ π/n^2 ,   n=1 000 000,…,1,   ∑_(1 000 000)▒〖=0,   ∑_0▒〖=S.〗〗〗



```python
import math
N = 1_000_000

sForward = 0.0 
for n in range(1, N + 1): #прямое суммирование n + 2
    sForward += math.pi / (n * n)

sBackward = 0.0
for n in range(N, 0, -1): #обратное суммирование n - 1
    sBackward += math.pi / (n * n)

abs_err = abs(sForward - sBackward)
rel_err = abs_err / abs(sBackward)

sForward, sBackward, abs_err, rel_err
```


КОНСПЕКТ: ПОГРЕШНОСТИ + ПОЧЕМУ ОБРАТНОЕ СУММИРОВАНИЕ ТОЧНЕЕ
	1.	ОТКУДА БЕРУТСЯ ПОГРЕШНОСТИ (главные типы)

	•	погрешность модели (мы описали реальность формулой, но формула может быть приближением)
	•	погрешность исходных данных (измерения, шум, округления во входных)
	•	погрешность метода (мы используем приближенный метод, например частичная сумма вместо бесконечной)
	•	вычислительная погрешность (округление при операциях на компе, float)

	2.	АБСОЛЮТНАЯ И ОТНОСИТЕЛЬНАЯ ПОГРЕШНОСТЬ

	•	абсолютная = насколько в лоб отличается число от эталона
abs_err = |x - x_exact|
	•	относительная = абсолютная, но поделили на размер эталона (процентик/доля)
rel_err = abs_err / |x_exact|
в задании эталоном приняли обратное суммирование, поэтому делим на него

	3.	ЧТО ТАКОЕ FLOAT И ПОЧЕМУ ОН ВРЕТ
float хранит число не “как бесконечные цифры”, а как:

	•	знак
	•	мантисса (ограниченное количество значащих бит/цифр)
	•	порядок/экспонента (типа * 10^k, но в двоичном виде)

важно: мантисса конечная. значит после каждой операции результат округляется. всегда.
	4.	ГЛАВНАЯ ПРИЧИНА: СЛОЖЕНИЕ В FLOAT НЕ АССОЦИАТИВНО
в математике (a+b)+c = a+(b+c)
в float это может быть НЕ равно, потому что округление происходит после каждого сложения
	5.	ПОЧЕМУ ОБРАТНОЕ СУММИРОВАНИЕ ТОЧНЕЕ (то, что тебя спросили)
идея простая: когда текущая сумма уже большая, прибавление очень маленького числа может вообще НИЧЕГО не изменить

почему так:
	•	у float есть “шаг” (минимальная добавка), которую он способен различить при данном масштабе числа
	•	чем больше по модулю число, тем крупнее этот шаг (потому что мантисса фиксированной длины, а порядок растёт)
	•	если добавка меньше этого шага, она “не помещается” в мантиссу и после округления пропадает

то есть реально может получиться:
S + tiny = S   (в float)
потому что tiny слишком мелкий на фоне S
	6.	КАК ЭТО СВЯЗАНО С ПОРЯДКОМ СУММИРОВАНИЯ
прямое суммирование (1 -> N):

	•	сначала добавляешь крупные члены, сумма быстро становится “большой”
	•	в конце начинаешь добавлять очень мелкие члены (типа 1/n^2 при n=1e6 это супер мелочь)
	•	и часть этих мелких может тупо теряться при округлении (поглощение)

обратное суммирование (N -> 1):
	•	начинаешь с самых маленьких членов
	•	они все одного порядка величины, складываются “нормально” и не исчезают
	•	сумма растёт постепенно, и только потом ты добавляешь крупные члены
итог: меньше потерь хвостов => точнее результат

	7.	КАК Я ПОНЯЛ ОБЪЯСНЕНИЕ ПРЕПОДА (как мне пизды дали)
“у float мантисса ограниченная и она заполняется значащими цифрами. когда сумма уже условно типа *10^6 (масштаб растёт), мелкие добавки перестают влезать в мантиссу и отсекаются, потому что порядок забирает разряды. поэтому при прямом суммировании хвосты теряются, а при обратном они успевают сложиться пока масштаб маленький”
	8.	КОРОТКИЙ ПРИМЕР (если препод опять будет душить)
есть классика:
1e16 + 1 = 1e16 (в float)  -> единица пропала
поэтому (1e16 + 1) - 1e16 = 0
а математически должно быть 1
это прям демонстрация “мантисса забита, мелочь не влезла”
	9.	ВЫВОД ПО ЗАДАЧЕ

	•	forward и backward отличаются не потому что формула разная, а потому что float округляет и порядок сложения важен
	•	backward обычно ближе к “истине” (в рамках вычисления на double), потому что меньше потери мел
