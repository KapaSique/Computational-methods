# Задача 8: Решение нелинейных уравнений
---

**Задача:** Методом простой итерации найти ширину функции $f(x) = x \cdot e^{-x}$, $x > 0$ на полувысоте с точностью $\varepsilon = 10^{-2}$.

### План решения

1. Найти максимум функции $f(x)$
2. Вычислить значение на полувысоте: $f_{1/2} = \frac{f_{\max}}{2}$
3. Решить уравнение $f(x) = f_{1/2}$ методом простой итерации для двух корней
4. Вычислить ширину: $\Delta x = x_2 - x_1$

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Шаг 1: Анализ функции и поиск максимума

```python
def f(x):
    """Функция f(x) = x * e^(-x)"""
    return x * np.exp(-x)

def f_derivative(x):
    """Производная f'(x) = e^(-x) - x*e^(-x) = e^(-x)(1 - x)"""
    return np.exp(-x) * (1 - x)

# Находим максимум: f'(x) = 0 => 1 - x = 0 => x = 1
x_max = 1.0
f_max = f(x_max)
f_half = f_max / 2

print(f"Максимум функции:")
print(f"  x_max = {x_max}")
print(f"  f_max = {f_max:.6f}")
print(f"  f_half = {f_half:.6f}")
```

### Визуализация функции

```python
x = np.linspace(0.01, 5, 500)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = x·e^(-x)')
plt.axhline(y=f_max, color='r', linestyle='--', label=f'f_max = {f_max:.4f}')
plt.axhline(y=f_half, color='g', linestyle='--', label=f'f_half = {f_half:.4f}')
plt.axvline(x=x_max, color='r', linestyle=':', alpha=0.5)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Функция f(x) = x·e^(-x)', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(0, 5)
plt.ylim(0, 0.4)
plt.show()
```

### Шаг 2: Метод простой итерации

Нужно решить уравнение: $x \cdot e^{-x} = f_{1/2}$

Преобразуем к виду $x = \phi(x)$:

**Для левого корня** ($x < 1$): $x = -\ln\left(\frac{f_{1/2}}{x}\right)$

**Для правого корня** ($x > 1$): используем другую форму итерации

```python
def simple_iteration(phi, x0, eps=1e-2, max_iter=1000):
    """
    Метод простой итерации для решения x = phi(x)
    
    Параметры:
    phi : итерационная функция
    x0 : начальное приближение
    eps : точность
    max_iter : максимальное число итераций
    
    Возвращает:
    x : найденный корень
    iterations : число итераций
    """
    x = x0
    for i in range(max_iter):
        x_new = phi(x)
        if abs(x_new - x) < eps:
            return x_new, i + 1
        x = x_new
    return x, max_iter
```

### Шаг 3: Поиск левого корня (x < 1)

```python
# Для левого корня: из x*e^(-x) = f_half => e^(-x) = f_half/x => -x = ln(f_half/x) => x = -ln(f_half/x)
def phi_left(x):
    return -np.log(f_half / x)

# Начальное приближение для левого корня
x0_left = 0.3
x_left, iter_left = simple_iteration(phi_left, x0_left, eps=1e-2)

print(f"Левый корень:")
print(f"  x_left = {x_left:.6f}")
print(f"  f(x_left) = {f(x_left):.6f}")
print(f"  Итераций: {iter_left}")
print(f"  Проверка: |f(x_left) - f_half| = {abs(f(x_left) - f_half):.2e}")
```

### Шаг 4: Поиск правого корня (x > 1)

```python
# Для правого корня используем ту же итерационную функцию
def phi_right(x):
    return -np.log(f_half / x)

# Начальное приближение для правого корня
x0_right = 3.0
x_right, iter_right = simple_iteration(phi_right, x0_right, eps=1e-2)

print(f"Правый корень:")
print(f"  x_right = {x_right:.6f}")
print(f"  f(x_right) = {f(x_right):.6f}")
print(f"  Итераций: {iter_right}")
print(f"  Проверка: |f(x_right) - f_half| = {abs(f(x_right) - f_half):.2e}")
```

### Шаг 5: Вычисление ширины на полувысоте

```python
width = x_right - x_left

print("="*60)
print("РЕЗУЛЬТАТ")
print("="*60)
print(f"Ширина функции на полувысоте: Δx = {width:.6f}")
print(f"Левая граница:  x₁ = {x_left:.6f}")
print(f"Правая граница: x₂ = {x_right:.6f}")
print("="*60)
```

### Визуализация результата

```python
x = np.linspace(0.01, 5, 500)
y = f(x)

plt.figure(figsize=(12, 7))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = x·e^(-x)')
plt.axhline(y=f_max, color='r', linestyle='--', alpha=0.5, label=f'f_max = {f_max:.4f}')
plt.axhline(y=f_half, color='g', linestyle='--', linewidth=2, label=f'f_half = {f_half:.4f}')

# Отмечаем найденные корни
plt.plot(x_left, f(x_left), 'ro', markersize=10, label=f'x₁ = {x_left:.4f}')
plt.plot(x_right, f(x_right), 'ro', markersize=10, label=f'x₂ = {x_right:.4f}')

# Показываем ширину
plt.plot([x_left, x_right], [f_half, f_half], 'r-', linewidth=3, alpha=0.7)
plt.annotate('', xy=(x_right, f_half), xytext=(x_left, f_half),
            arrowprops=dict(arrowstyle='<->', color='red', lw=2))
plt.text((x_left + x_right)/2, f_half + 0.02, f'Δx = {width:.4f}', 
         ha='center', fontsize=12, color='red', fontweight='bold')

plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Ширина функции на полувысоте', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.xlim(0, 5)
plt.ylim(0, 0.4)
plt.show()
```

### Конспект: Метод простой итерации

**Идея:** Преобразовать уравнение $f(x) = 0$ к виду $x = \phi(x)$ и строить последовательность:

$$x_{n+1} = \phi(x_n), \quad n = 0, 1, 2, \ldots$$

**Условие сходимости:** Если $|\phi'(x)| < 1$ в окрестности корня, то метод сходится.

**Критерий остановки:** $|x_{n+1} - x_n| < \varepsilon$

**Преимущества:**
- Простота реализации
- Не требует вычисления производной

**Недостатки:**
- Может не сходиться при неудачном выборе $\phi(x)$
- Медленная сходимость по сравнению с методом Ньютона
